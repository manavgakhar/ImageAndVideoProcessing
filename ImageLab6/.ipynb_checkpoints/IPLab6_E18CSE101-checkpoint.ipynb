```python
from scipy import misc, signal
from scipy import ndimage, misc
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
import cv2
print(np.max(image))
print(image.shape)
```

```python
def laplacian(image):
    v = image
    s = cv2.Laplacian(v, cv2.CV_16S, ksize=3)
    s = cv2.convertScaleAbs(s)
    cv2.imshow('nier',s)
    return s
```

```python
#Laplacian
plt.title('Laplacian')
plt.imshow(laplacian(image),cmap='gray')
```

```python
def gauss_laplacian(image):
    v = image
    g= cv2.GaussianBlur(v,(5,5),cv2.BORDER_DEFAULT)
    s = cv2.Laplacian(g, cv2.CV_16S, ksize=3)
    s = cv2.convertScaleAbs(s)
    cv2.imshow('nier',s)
    return s
```

```python
#Gaussian then Laplacian Filter
plt.title('Gaussian then Laplacian')
plt.imshow(gauss_laplacian(image),cmap='gray')
```

## Gaussian smoothing then laplacian shafrpening gives a better result as image is clearer and has less noise

```python
lapl= laplacian(image)
```

```python
#adding original image
```

```python
og_lap= image+lapl
```

```python
plt.title('Laplacian Filter + Original Image')
plt.imshow(og_lap,cmap='gray')
```

```python
sob= sobel(image)
```

```python
sob_smoothed= cv2.GaussianBlur(sob,(5,5),cv2.BORDER_DEFAULT)
```

```python
plt.title('Sobel + Gaussian smoothing')
plt.imshow(sob_smoothed,cmap='gray')
```

```python
#product of C and E
prod = sob_smoothed*og_lap
```

```python
plt.title('Product of c and e')
plt.imshow(prod,cmap='gray')
```

```python
#sum of a and f
summ= image+prod
```

```python
plt.title('Sum of a and f')
plt.imshow(summ,cmap='gray')
```
