```python
import matplotlib.pyplot as plt
from skimage.transform import warp
from skimage import data
from skimage.transform import SimilarityTransform
import cv2
```

```python
img= cv2.imread('tajmahal.jpg')
```

## Question 1

```python
plt.imshow(img)
```

```python
a=SimilarityTransform(translation=(30,-10))
```

```python
imgg= warp(img,a)
```

```python
##
```

```python
plt.title("(30,-10)")
plt.imshow(imgg)
```

```python
a=SimilarityTransform(translation=(30,10))
imgg= warp(img,a)
plt.title("(30,10)")
plt.imshow(imgg)
```

```python
a=SimilarityTransform(scale=0.8,rotation=1,translation=(30,10))
imgg= warp(img,a)
plt.title("Scale,Rotation,Translation")
plt.imshow(imgg)
```
