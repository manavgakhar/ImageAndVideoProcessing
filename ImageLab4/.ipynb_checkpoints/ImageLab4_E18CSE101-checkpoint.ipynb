```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_= cv2.imread("Images1/cameraman.jpg")
```

```python
img_.shape
```

```python
img_gray= cv2.cvtColor(img_,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
img_gray.shape
```

```python
f_min= np.min(np.array(img_gray))
```

```python
f_max= np.max(np.array(img_gray))
```

```python
f_max
```

```python
levels= [64,32,16,4]
```

## Plotting with different gray scale levels

```python
levels= [64,32,16,4]
for i in levels:
    rangee= (f_max-f_min)/i
    img_new= img_gray//rangee
    plt.imshow(img_new,cmap='gray')
    plt.show()
```

### Using in built function

```python
from PIL import Image
```

```python
img_gray= Image.fromarray(img_gray)
```

```python
img_gray.quantize(colors=4)
```

```python
for i in levels:
    bruh=img_gray.quantize(colors=i)
    plt.imshow(bruh)
    plt.show()
```

```python
ms=[]
for i in levels:
    rangee= (f_max-f_min)/i
    img_new= img_gray//rangee
    mse= np.mean((img_new-img_gray)**2)
    ms.append(mse)
```

```python
import seaborn as sns
```

```python
sns.barplot(levels,ms)
plt.xlabel("Levels")
plt.ylabel("MSE")
```

## As can be seen above, as L decreases, image quality decreases and MSE increases

## Question 2

```python
img1 = cv2.imread("Images1/parrot.jpg")
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
bw_img = cv2.threshold(img1,127,255,cv2.THRESH_BINARY)[1]
```

```python
plt.imshow(bw_img,cmap='gray')
```

```python
#Euclidean distance
euc = cv2.distanceTransform(bw_img,cv2.DIST_L1, 3)
plt.imshow(euc,cmap='gray')
```

```python
#city- block distance
cb = cv2.distanceTransform(bw_img,cv2.DIST_L2, 3)
plt.imshow(cb,cmap='gray')
```

```python
#chessboard distance
chess = cv2.distanceTransform(bw_img,cv2.DIST_C, 3)
plt.imshow(chess,cmap='gray')
```

# 3x3

```python
import cv2 
import numpy as np 
   
      
# Read the image 
img = cv2.imread('Images1/bird.png', 0) 
  
# Obtain number of rows and columns  
# of the image 
m, n = img.shape 
   
# Develop Averaging filter(3, 3) mask 
mask = np.ones([3, 3], dtype = int) 
mask = mask / 9
   
# Convolve the 3X3 mask over the image  
img_new = np.zeros([m, n]) 
  
for i in range(1, m-1): 
    for j in range(1, n-1): 
        temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2] 
         
        img_new[i, j]= temp 
          
img_new = img_new.astype(np.uint8) 
cv2.imwrite('blurred.png', img_new) 
```

```python
imge = cv2.imread('blurred.png', 0) 
pyplot.imshow(imge)
```

# 5x5
