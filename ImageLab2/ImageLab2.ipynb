## RGB to Grayscale using only numpy and matplotlib

```python
image_1= mpimg.imread("image2.jpg")
```

```python
plt.imshow(image_1)
```

## A grayscale image is one in which only the information about intensity of light is stored in every pixel.

## - To convert RGB to Grayscale, we need to take the weighted sum of each channel 

## - This weighted sum can be taken using numpy easily

```python
#defining the weights for each channel
weights= [0.2989, 0.5870, 0.1140]
```

```python
#now we take dot product
image_1_gray= np.dot(image_1[...,:3], weights)
```

```python
fig,axes= plt.subplots(1,2)
axes[0].imshow(image_1)
axes[1].imshow(image_1_gray,cmap='gray')
```

## Question 4

```python
## Image interpolation using opencv
```

```python
img3= mpimg.imread("image3.jpg")
```

```python
img3.shape
```

```python
img3=img3[50:100,50:100,:]
```

```python
plt.imshow(img3)
```

## Out of the three methods used below, nearest neighbour gives the most pixelated image and bilinear gives the best image

```python
#bilinear
plt.imshow(cv2.resize(img3,None, fx = 150, fy = 150, interpolation = cv2.INTER_LINEAR))
```

```python
#nearest neighbour
plt.imshow(cv2.resize(img3,None, fx = 150, fy = 150, interpolation = cv2.INTER_NEAREST))
```

```python
#cubic
plt.imshow(cv2.resize(img3,None, fx = 150, fy = 150, interpolation = cv2.cv2.INTER_CUBIC))
```
